```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargamos el archivo CSV
df = pd.read_csv('peliculas - peliculas.csv')

# Mostrar las primeras 5 filas solo para verificar q si lo esta leyendo
df.head()
```

```python
# Estos seran los generos que voy a filtrar
generos_interes = ['Terror', 'Romance']

# Filtramos el DataFrame
df_filtrado = df[df['genero'].isin(generos_interes)]

# Y confirmamos los registros que quedan
print(f"Registros después de filtrar: {len(df_filtrado)}")
df_filtrado.head()
```

```python
# Aqui se agrupoa por ano y se saca la media usando pandas
promedio_por_anio = df_filtrado.groupby('anio')['calificacion'].mean()

# Mostramos los primeros resultados
promedio_por_anio.head(26) # Declaro el numero 26 para que muestre todos los resultados
```

```python



#Creamos las graficas
plt.figure(figsize=(10, 6))
promedio_por_anio.plot(kind='line', marker='o', color='purple') # Anadimos marcadores, tipo de linea y un color bonito

plt.title('Promedio de Calificación por Año (Thriller y Ciencia Ficción)')
plt.xlabel('Año')
plt.ylabel('Calificación Promedio')
plt.grid(True, linestyle='--', alpha=0.7) #El grid lo anadimos solo por estetica, para que se vea "mas profesional"

plt.show()
```

```python
# Grafico de barras para el ano 2000 (Extra de la tarea)

# Filtramos para el ano 2000
peliculas_2000 = df[df['anio'] == 2000]


anio2000 = peliculas_2000.groupby('genero')['calificacion'].mean().sort_values()

plt.figure(figsize=(10, 6))
anio2000.plot(kind='bar', color='red')

plt.title('Promedio de Calificación por Género (Año 2000)')
plt.xlabel('Género')
plt.ylabel('Calificación Promedio')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Esto solo fue copia y pega del grafico de arriba


plt.show()
```
